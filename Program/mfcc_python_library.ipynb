```python
from python_speech_features import mfcc
from python_speech_features import delta
from python_speech_features import logfbank
from python_speech_features import get_filterbanks
from python_speech_features import fbank
import scipy.io.wavfile as wav
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import cm
```

```python
audio_path1 = "../Dataset/01.Ha\'/Ha\'_M1_1.wav"
# audio_path1 = "../Dataset/01.Ha\'/Ha\'_RY1_1.wav"
```

```python
(rate,sig) = wav.read(audio_path1)
```

```python
print(rate, sig.shape)
```

```python
mfcc_feat = mfcc(sig,
                samplerate=16000,
                winlen=0.025,
                winstep=0.01,
                numcep=13,
                nfilt=20,
                preemph=0.97,
                ceplifter=22,
                winfunc=np.hamming)
        
```

```python
# mfcc_feat = np.mean(mfcc_feat, axis=0)
```

```python
print(mfcc_feat)
print(type(mfcc_feat), mfcc_feat.shape)
```

```python
plt.plot(mfcc_feat)
plt.show()
```

```python
fig, ax = plt.subplots()
fig.patch.set_facecolor('white')
mfcc_data= np.swapaxes(mfcc_feat, 0 ,1)
cax = ax.imshow(mfcc_data, interpolation='nearest', cmap=cm.coolwarm, origin='lower', aspect='auto')
ax.set_title('MFCC')
#Showing mfcc_data
plt.show()
```

```python
fbank_feat, energy_feat = fbank(sig, samplerate=16000, winlen=0.025, winstep=0.01, nfilt=20, nfft=512, winfunc=np.hamming)
```

```python
plt.plot(fbank_feat)
plt.show()
```

```python
logfbank_feat= np.log(fbank_feat)
```

```python
plt.plot(logfbank_feat)
plt.show()
```
