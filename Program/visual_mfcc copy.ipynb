```python
import librosa
import numpy as np
import noisereduce as nr
import matplotlib.pyplot as plt

def preprocessing(audio):
    y , sr = librosa.load(audio, sr=44100)
    y, _ = librosa.effects.trim(y, top_db=20)
    # y = librosa.util.fix_length(y, size=600000, mode = 'constant')
    y = nr.reduce_noise(y, sr)
    return y, sr

def extraction(audio):
    y , sr = preprocessing(audio)
    y = librosa.effects.preemphasis(y)
    mfccs = librosa.feature.mfcc(y=y,
                                 sr=sr,
                                 n_mfcc=13,
                                 window='hamming',
                                 win_length=1024,
                                 htk = False,
                                 hop_length=512,
                                 n_fft=1024,
                                 norm='ortho',
                                 dct_type=2,
                                 n_mels=128)
    mfccs = librosa.util.normalize(mfccs, axis=1)
    return mfccs
```

```python
audio1 = "../Dataset/01.Ha\'/Ha\'_M1_11.wav"
audio2 = "../Dataset/01.Ha\'/Ha\'_M1_12.wav"
audio3 = "../Dataset/01.Ha\'/Ha\'_M1_13.wav"

mfccs1 = extraction(audio1)
mfccs2 = extraction(audio2)
mfccs3 = extraction(audio3)

# Plot MFCC dalam satu baris
fig = plt.figure(figsize=(17, 8), facecolor='white')  # Menambahkan latar belakang putih
plt.subplot(2, 3, 1)
librosa.display.specshow(mfccs1, x_axis='time')
plt.colorbar()
plt.title(audio1)
plt.subplot(2, 3, 2)
librosa.display.specshow(mfccs2, x_axis='time')
plt.colorbar()
plt.title(audio2)
plt.subplot(2, 3, 3)
librosa.display.specshow(mfccs3, x_axis='time')
plt.colorbar()
plt.title(audio3)
plt.subplot(2, 3, 4)
librosa.display.specshow(mfccs1, x_axis='time', cmap='viridis')
plt.colorbar()
plt.title(audio1)
plt.subplot(2, 3, 5)
librosa.display.specshow(mfccs2, x_axis='time', cmap='viridis')
plt.colorbar()
plt.title(audio2)
plt.subplot(2, 3, 6)
librosa.display.specshow(mfccs3, x_axis='time', cmap='viridis')
plt.colorbar()
plt.title(audio3)
plt.tight_layout()

plt.savefig('../new_hasil_gambar/norm_ha_mfcc_visual.png')

plt.show()
```
